Prereceptoral filtering
=======================

Before incident light hits retinal photoreceptors it first travels through the crystalline lens and other ocular media, which act as prereceptoral filters. 


Lens yellowing with age
-----------------------


<img src="../../img/lensesAge.jpg" width="200" height="200" />     

The lens contains a yellow pigment which accumulates with age due to absorption of UV light. 





```python
from pysilsub import CIE

CIE.get_CIEPO06_optical_density()['D_ocul_1'].plot(
    xlabel='Wavelength (nm)', ylabel='Lens pigment optical density')
```
