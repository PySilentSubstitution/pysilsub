Package data
============

*PySilSub* ships with calibration and configuration data for various multiprimary stimulation devices that we have used for silent substitution. This means you can experiment with the package without having to calibrate and configure your own stimulation device!

```python
from pysilsub.device import StimulationDevice

StimulationDevice.show_package_data(verbose=False)
```

One can easily create a ``StimulationDevice`` or ``SilentSubstitutionProblem`` from example package data:

```python
from pprint import pprint

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

device = StimulationDevice.from_package_data('STLAB_Oxford')

device.plot_calibration_spds(ax=axs[0])
device.plot_gamut(ax=axs[1])
pprint(device.config)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

device = StimulationDevice.from_package_data('VirtualSky')

device.plot_calibration_spds(ax=axs[0])
device.plot_gamut(ax=axs[1])
pprint(device.config)
```
