```python
import pprint

import matplotlib.pyplot as plt

from pysilsub.device import StimulationDevice
```

```python
sd = StimulationDevice.from_json('../data/STLAB_1_York.json')

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sd.plot_spds(ax=axs[0])
sd.plot_gamut(ax=axs[1])

pprint.pprint(sd.config)
```
