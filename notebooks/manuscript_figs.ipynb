```python
import matplotlib.pyplot as plt

from pysilsub.device import StimulationDevice
```

```python
device = StimulationDevice.from_package_data('STLAB_Oxford')

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

device.predict_primary_spd(3, .5).plot(ax=axs[0], c='b')
axs[0].set_title('device.predict_primary_spd(0, .5)\n')
device.predict_multiprimary_spd([.5] * device.nprimaries).plot(ax=axs[1], c='k')
axs[1].set_title('spec = [.5] * device.nprimaries\ndevice.predict_multiprimary_spd(spec)\n')
device.predict_multiprimary_aopic([.5] * device.nprimaries).plot(kind='bar', color=device.aopic_colors.values())
axs[2].set_title('device.predict_multiprimary_aopic()\n')
```
